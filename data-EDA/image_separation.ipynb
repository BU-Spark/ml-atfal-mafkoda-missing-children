# Create Databse & Image Separation on Multiple Faces Images

```python
# get the path/directory
# create the database for missing people (database)
# iterate over the raw image (data), and create folder such as person1, person2, ...
# We put the separated images into the same person folder

# please change to your desired directory
folder_dir = "atfalmafkoda_unzip/missing_children/"
person_index = 1
for path in os.listdir(folder_dir):
    img = matplotlib.image.imread(folder_dir+path)
    face_detector = mtcnn.MTCNN()
    face_roi = face_detector.detect_faces(img)
    if len(face_roi)>=2:
        tiles = image_slicer.slice(folder_dir+path, len(face_roi), save=False)
        # please change to your desired directory
        new_path = "/projectnb/sparkgrp/ml-atfal-mafkoda-grp/atfalmafkoda_unzip/database/person" + str(person_index)
        os.makedirs(new_path)
        image_slicer.save_tiles(tiles, prefix='slice', format='JPEG', directory=new_path)
    else:
        # please change to your desired directory
        new_path = "/projectnb/sparkgrp/ml-atfal-mafkoda-grp/atfalmafkoda_unzip/database/person" + str(person_index)
        os.makedirs(new_path)
        im = Image.fromarray(img)
        im.save(f"{new_path}/slice_01_01.png")
    person_index += 1
```
